### 2.1 Webscraping data from Fbref

### 2.1.1 Webscraping Player data

The following function extracts the commented our player table from fbref.com

```python
def downloadPlayerTable(url, filename):

    response = requests.get(url, headers = headers)
    
    #check if the page can be scraped
    if response.status_code == 200:
        
        soup = bs(response.text, "html.parser")
        #Find the table that is commented out
        comments = soup.find_all(string=lambda text: isinstance(text, Comment))

        tables = []
        for each_line in comments:
            if 'table' in each_line:
                try:
                    #use extend instead of list.append so we don't have a 2d dataframe
                    tables.extend(pd.read_html(each_line))
                except:
                    continue

        df = pd.concat(tables, ignore_index = True)
        df.to_csv(f'{filename}.csv', index=False)
        print(f'Successfully created {filename}.csv')
    else:
        print(f'{response.status_code}: Error connecting to site')
```

### 2.1.2 Webscraping team data 
The following function extracts data from the team table

```python
def downloadTeamTable(url, filename):
    
    response = requests.get(url, headers = headers)
    
    #check if the page can be scraped
    if response.status_code == 200:
        
        soup = bs(response.text, "html.parser")
        #find the table from the page
        table = soup.find('table', class_ = "stats_table")
        
        if table:
            #read the table data to a pandas dataframe and save to csv file
            df = pd.read_html(str(table))[0] 
            df.to_csv(f'{filename}.csv', index=False)
            print(f'Successfully created {filename}.csv')
        else:
            print(f'{response.status_code}: Error connecting to site')
        
```

### 3.1 Cleaning PlayerStats
The playerstats and miscstats tables require the same type of cleaning which involves removing the leftmost column from the dataframe, removing null values, removing duplicate headers, filtering by the top 10 teams, removing the last column from the table, and changing data to correct datatypes. Instead of doing this for both tables, I will instead create a function and get the clean table and display the datatypes and a description of both tables. This has been done since it will be much faster and make the code look cleaner. We also have to filter all datasets by the top 10 teams.

### Function to clean both tables

```python
def cleanPlayerData (df):
    #remove leftmost column from dataframe
    df = df.iloc[:, 1:]

    #drop matches column
    df = df.drop(columns = ['Matches'])

    #remove null values
    df = df.dropna()

    #remove duplicate headers
    df = df[df.iloc[:, 0] != df.columns[0]]

    #Create regex pattern with OR conditions
    pattern = '|'.join(Teams)

    #Return true or false based on if player is in top 10 team
    topTen = df["Squad"].str.contains(pattern, case = False)

    #Show new dataset with only players from top 10 teams
    df = df[topTen]
    #Renumber the table
    df.reset_index(drop = True, inplace = True)
    # #Convert all non string values to appropriate datatype
    df = df.apply(pd.to_numeric, errors= 'ignore')
    
    return df
```

### 4.2 Goals scored vs xG
Expected goals (xG) is matric that's intended to measure the probability of a shot resulting in a goal. The purpose of this is to show when a player should be expected to score from a particular opportunity, by basically rating how good a goal-scoring opportunity is. This provides an additional perspective on how a team could secure a league victory, evident when their high Expected Goals (xG) surpasses the actual goals scored. While a substantial xG already suggests the team's anticipated ability to score plenty of goals, surpassing the expected goal count indicates that the team is not only meeting but exceeding expectations—a clear sign of overperformance and a potent attacking threat.

```python
# Sample target teams
target_teams = ['Manchester City', 'Arsenal', 'Liverpool', 'Manchester Utd']

# Split scores using en dash
matchdayResults['HomeGoals'] = matchdayResults['Score'].str.split('\u2013', expand=True)[0]
matchdayResults['AwayGoals'] = matchdayResults['Score'].str.split('\u2013', expand=True)[1]

matchdayResults['HomeGoals'] = pd.to_numeric(matchdayResults['HomeGoals'])
matchdayResults['AwayGoals'] = pd.to_numeric(matchdayResults['AwayGoals'])

for i, target_team in enumerate(target_teams, 1):
    xg_name = target_team + " xG"

    team_matches = matchdayResults[(matchdayResults['Home'] == target_team) | (matchdayResults['Away'] == target_team)].copy()
    # Add scores and xG values based on home/away status
    team_matches.loc[team_matches['Home'] == target_team, 'TotalGoals'] = matchdayResults['HomeGoals']
    team_matches.loc[team_matches['Away'] == target_team, 'TotalGoals'] = matchdayResults['AwayGoals']
    
    team_matches.loc[team_matches['Home'] == target_team, 'TotalxG'] = matchdayResults['xG']
    team_matches.loc[team_matches['Away'] == target_team, 'TotalxG'] = matchdayResults['xG.1']
    team_matches['CumulativeTotalGoals'] = team_matches['TotalGoals'].cumsum()
    team_matches['CumulativeTotalxG'] = team_matches['TotalxG'].cumsum()

    # Plotting
    _ = plt.plot(team_matches.index, team_matches['CumulativeTotalGoals'], marker='o', linestyle='-', color='r', label='Goals')
    _ = plt.plot(team_matches.index, team_matches['CumulativeTotalxG'], marker='o', linestyle='-', color='y', label='Expected Goals')
    _ = plt.title(f'Cumulative Total Goals Over Time for {target_team}')
    _ = plt.xlabel('Date')
    _ = plt.ylabel('Cumulative Total Goals')
    _ = plt.grid(True)
    plt.show();
print("Red line- Goals scored")
print("Yellow line- Expected Goals")
```

Manchester City exceeded a relatively high Expected Goals (xG) of 79, showcasing their prowess as an attacking force with an impressive goal tally. This suggests not only a high-scoring team but also one capable of converting chances into quality goals, reflecting a dominance in possession and game control.

Contrastingly, Arsenal achieved a substantial goal count despite a comparatively lower xG than Manchester City. This indicates clinical finishing, highlighting the team's ability to capitalize on goal-scoring opportunities even when they may not be of high quality. Factors like goalkeeper errors or a counter-attacking style could contribute to this efficiency.

Liverpool's goal tally closely aligned with their xG, suggesting they converted their scoring chances efficiently. In contrast, Manchester United scored fewer goals than their xG, indicating missed opportunities or strong opposition goalkeeping preventing them from capitalizing on quality chances. Additional factors, such as hitting the post or defensive interventions, may also have influenced their goal-scoring performance.

### 4.4 Win Percentage by day
This can provide insights into how the team performed throughout the duration of the league. We can analyse the team's concistency by examining win percentage by day.

```python
#Filter based on manchester city
team_matches = matchdayResults[(matchdayResults['Home'] == 'Manchester City') | (matchdayResults['Away'] == 'Manchester City')].copy()

# Split scores using en dash
team_matches['HomeGoals'] = team_matches['Score'].str.split('\u2013', expand=True)[0]
team_matches['AwayGoals'] = team_matches['Score'].str.split('\u2013', expand=True)[1]
# Create a new column 'manchester_city_wins' to store the information
team_matches['manchester_city_wins'] = ''

# Check if Manchester City is the home or away team and if they won
team_matches.loc[(team_matches['Home'] == 'Manchester City') & (team_matches['HomeGoals'] > team_matches['AwayGoals']), 'manchester_city_wins'] = 'Yes'
team_matches.loc[(team_matches['Away'] == 'Manchester City') & (team_matches['AwayGoals'] > team_matches['HomeGoals']), 'manchester_city_wins'] = 'Yes'
team_matches.loc[(team_matches['Home'] == 'Manchester City') & (team_matches['HomeGoals'] <= team_matches['AwayGoals']), 'manchester_city_wins'] = 'No'
team_matches.loc[(team_matches['Away'] == 'Manchester City') & (team_matches['AwayGoals'] <= team_matches['HomeGoals']), 'manchester_city_wins'] = 'No'

# Count the number of games played on each day
games_played = team_matches.groupby('Day')['manchester_city_wins'].count().reset_index(name='games_played')

win_percentages = []
unique_days = team_matches['Day'].unique()

for day in unique_days:
    day_df = team_matches[team_matches['Day'] == day]
    #Calculate the amount of Yes for each day and get a percentage
    win_percentage = (day_df['manchester_city_wins'].eq('Yes').sum() / day_df.shape[0]) * 100
    win_percentages.append(win_percentage)

# Subplot with layour
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 7))

# Create first graph with win percentage each day
axes[0].bar(unique_days, win_percentages, color='blue', label="Games won Percentage")
axes[0].set_xlabel('Day')
axes[0].set_ylabel('Win Percentage')
axes[0].set_title('Manchester City Win Percentage on Each Day')
axes[0].set_ylim(0, 100)  # Set y-axis limits to 0-100 for percentage


# Create second graph for number of games played on each day
axes[1].bar(games_played['Day'], games_played['games_played'], color='green', alpha = 0.5, label='Games Played')
axes[1].set_ylabel('Number of Games Played', color='green')
axes[1].set_xlabel('Day')
axes[1].set_title('Manchester City Number of games played on Each Day')
axes[1].tick_params('y', colors='green')

plt.legend()
plt.show();
```

The provided data illustrates Manchester City's varied win rates across different days: approximately 62% on Sunday, 75% on Saturday, 80% on Wednesday, and a perfect 100% on Thursday. Notably, while Wednesday and Thursday boast 100% and 80% win rates, respectively, a closer look at the corresponding chart reveals that these are the days with the fewest games played, numbering 6 and 2, respectively. In contrast, Saturday stands out with 19 games and a 75% win rate, suggesting that Manchester City tends to perform exceptionally well on Saturdays. The team's impeccable 100% win rate on Thursday raises the possibility that they faced weaker opponents on that day. Worth noting is that both opponents on Thursday were Chelsea, a team that did not place in the top 10, and Spurs, which secured a relatively low position in the standings.

One cruical aspect into Pep's tactics is build up play with a lot of passing. He generally went with a 4 man midfield and a 3 man attack that ensured his team had a lot of posession.

PrgC- Progressive Carries which is carries that move the ball towards the opponent's goal line at least 10 yards from its furthest point in the last six passes or any carry into the penalty area.

PrgP - Progressive Passes  which is completed passes that move the ball towards the opponent's goal line at least 10 yard from its furthest point in the last six passes or any completed pass into the penalty area.

For the data below, we will put different combinations of players in different groups and get the mean, median and standard deviation of these groups which can give us some insight on the team's formation.

```python
#Gather all midfielders from manchester city
midfielders = playerStats[playerStats['Pos'].str.contains("MF")].copy()
midfielders = midfielders[midfielders['Squad'] == 'Manchester City']

midfielder_names = []
midfielder_names.extend(midfielders['Player'])

# Create a dictionary to store the lists with random player names
random_name_lists = {}
groups = {}
std_values = []

# Generate 5 lists of 5 random names each and assign them names
for i in range(5):
    list_name = f"List{i+1}"
    random_name_lists[list_name] = list(np.random.choice(midfielder_names, size=5, replace=False))
    
    
#Put the top 5 midfielders in a list
List6 = midfielders.nlargest(5, 'Starts')['Player'].tolist()
# Assign names to the dataframes

#Put the midfielder names into dataframes
for i in range(5):
    group_name = f'group{i+1}'
    
    #Initialize dataframe in the group
    groups[group_name] = pd.DataFrame()
    groups[group_name]['Player'] = random_name_lists[f'List{i+1}']
    
#Initialize dataframe for group 6
groups['group6'] = pd.DataFrame()
groups['group6']['Player'] = List6

#Get all player data from midfielders dataframe
for i in range(6):
    group_name = f'group{i+1}_df'
    group_name_nodf = f'group{i+1}'
    
    group_name = midfielders[midfielders['Player'].isin(groups[f'group{i+1}']['Player'])]
    
    mean_value = group_name['PrgP'].mean()
    std_value = group_name['PrgP'].std()
    std_values.append({'Group':group_name_nodf, 'Mean': mean_value, 'Std': std_value})
    
    
# Set up a 2x3 subplot grid
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))

# Flatten the axes for easy iteration
axes = axes.flatten()

# Display DataFrames in the subplot grid
for i in range(1, 7):
    group_name = f"group{i}"
    axes[i - 1].axis('off')  # Turn off axis for better appearance
    axes[i - 1].table(cellText=groups[group_name].values,
                     colLabels=groups[group_name].columns,
                     cellLoc='center', loc='center')
    axes[i - 1].set_title(f'Group {i}')

plt.show();
```
